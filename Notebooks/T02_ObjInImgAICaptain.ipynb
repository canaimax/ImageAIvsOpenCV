```python
detector = ObjectDetection()
```

```python
detector.setModelTypeAsYOLOv3()
```

```python
detector.setModelPath("/content/drive/MyDrive/colab/yolo.h5")
```

```python
detector.loadModel()
```

```python
detections = detector.detectObjectsFromImage(input_image="/content/drive/MyDrive/img77.jpg", output_image_path="imagenew.jpg", minimum_percentage_probability=30)
```
